Формулировка задачи:
  - Даны текст на русском языке и множество жанров, к к которому текст можно отнести. Необходимо определить жанровую принадлежность сюжетного текста кинофильма. Использование методов обработки больших данных должно установить связь между описанием сюжета фильма и его жанровой принадлежности.

Описание данных:
  - Для постановки задачи требуется ввести понятие "жанр". Жанр - группа произведений внутри рода, объединенная общими формальными, содержательными и функциональными признаками. Для определения жанра текста событий кинофильма необходимо прежде всего определить род самого сюжета, поскольку описываемый сюжет всегда проходит через призму восприятия зрителя или автора.

Откуда брать данные:
  - Сюжетную составляющую фильмов, как и множество жанров, возьмем с сайта ivi.ru. На данном сайте у каждого фильма, на его собственной странице имеется небольшое описание картины. Так как предоставленные там описания могут быть коротки, а объем данных недостаточно велик, это может повлиять на процент ошибок.

Алгоритмы:
  - Для начала рассмотрим распределение по жанрам и уберем редко встречающиеся жанры.
  - Перед началом работы текст будет обработан - произведем токенизацию. Для токенизации потребуется модификация библиотеки NLTK (Natural Language Toolkit) для работы с русским языком. Далее слова помещаются в словарь и сортируются по частоте использования в тексте.
  - Далее используется кросс-валидация, а также оцениваются модели линейных опорных векторов, логистической регрессии и тп
