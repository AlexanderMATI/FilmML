```python
import requests
from bs4 import BeautifulSoup
from dateutil.parser import parse
import concurrent.futures
import pandas as pd
from selenium.webdriver import Chrome
from selenium.webdriver import Firefox
import time
import multiprocessing
```

```python
MAX_THREADS = 50
```

```python
movie_title_arr = []
movie_year_arr = []
movie_genre_arr = []
movie_synopsis_arr =[]
```

```python
def getMovieTitle(header):
    try:
        return header[0].getText()
    except:
        return 'NA'

def getReleaseYear(muted_text):
    try:
        return muted_text.find_all("a",  {"class": "link parameters__link"})[0].getText()
    except:
        return 'NA'

def getGenre(muted_text):
    try:
        return muted_text.find_all("a",  {"class": "link parameters__link"})[2].getText()
    except:
        return 'NA'

def getSynopsys(synopsis):
    try:
        return synopsis.find_all("p")[0].getText()
    except:
        return 'NA'
```

```python
def main(main_url, collections_url):
    print(collections_url)

    driver.get(collections_url)
    time.sleep(5)
    response = driver.page_source

    soup = BeautifulSoup(response, 'html.parser')
    movies_list = soup.find_all("li", {"class": "gallery__item gallery__item_virtual"})

    for movie in movies_list:
        posterBlock = movie.find("a", {"class":  "nbl-slimPosterBlock"})
        if posterBlock is None:
          continue
        link = posterBlock.get('href')

        newResponse = requests.get(main_url + link)
        newSoup = BeautifulSoup(newResponse.text, 'html.parser')

        header = movie.find_all("div", {"class":  "nbl-slimPosterBlock__title"})
        
        muted_text = newSoup.find("div",{"class": "parameters__info"})
        
        movie_title =  getMovieTitle(header)
        if movie_title in movie_title_arr:
          continue
        else:
          movie_title_arr.append(movie_title)
        
          year = getReleaseYear(muted_text)
          movie_year_arr.append(year)
        
          genre = getGenre(muted_text)
          movie_genre_arr.append(genre)
        
          synopsis = getSynopsys(newSoup.find("div",{"class": "clause__text-inner"}))
          movie_synopsis_arr.append(synopsis)
```

```python
def parse_page(num):
  collections_url = f'https://www.ivi.ru/movies/all/page{num}?sort=ivi&rating_part=main&rating_model=ready'
  imageArr.append(collections_url)
  main(f'https://www.ivi.ru', collections_url)

imageArr = []
MAX_PAGE = 333

#первая страница
collections_url = f'https://www.ivi.ru/movies/all?sort=ivi&rating_part=main&rating_model=ready'
imageArr.append(collections_url)
main(f'https://www.ivi.ru', collections_url)

#остальные страницы
pool_obj = multiprocessing.Pool(6)
answer = pool_obj.map(parse_page, range(2, MAX_PAGE))

    
```

```python
def download_stories(story_urls):
    threads = min(MAX_THREADS, len(story_urls))
    with concurrent.futures.ThreadPoolExecutor(max_workers=threads) as executor:
        executor.map(main, story_urls)
```

```python
download_stories(imageArr)

movieDf = pd.DataFrame({
    "Title": movie_title_arr,
    "Release_Year": movie_year_arr,
    "Genre": movie_genre_arr,
    "Synopsis": movie_synopsis_arr,
})

print('--------- Download Complete CSV Formed --------')

movieDf.head()
```
